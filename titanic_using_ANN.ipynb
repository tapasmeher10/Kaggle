##Import Libraries

```python
import numpy as np
import pandas as pd
import tensorflow as tf
```

##Importing Dataset

```python
dataset=pd.read_csv("train.csv")
testset=pd.read_csv("test.csv")
```

#**Preprocessing the Data**


##Nan value

```python
dataset.columns[dataset.isna().any()]
```

```python
testset.columns[testset.isna().any()]
```

```python
dataset.Age=dataset.Age.fillna(dataset.Age.mean())
```

```python
dataset.dropna(inplace = True)
```

```python
dataset.columns[dataset.isna().any()]
```

```python
testset.Age=testset.Age.fillna(testset.Age.mean())
```

```python
testset.Fare=testset.Fare.fillna(testset.Fare.mean())
```

##split into x and y

```python
X=dataset.iloc[: ,3:-1].values
y=dataset.iloc[:,-1].values
```

```python
xt=testset.iloc[:,2:].values
```

```python
print(xt)
```

##encoding categorical data

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
X[:,1]=le.fit_transform(X[:,1])
xt[:,1]=le.fit_transform(xt[:,1])
print(X)
```

##**one hot encoding**


```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct= ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[5])],remainder='passthrough')
tct= ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[5])],remainder='passthrough')
X=np.array(ct.fit_transform(X))
xt=np.array(tct.fit_transform(xt))
```

```python
print(X)
```

##Feature scalling

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X=sc.fit_transform(X) 
```

```python
from sklearn.preprocessing import StandardScaler
sc1=StandardScaler()
xt=sc1.fit_transform(xt) 
```

##ANN modeling

```python
ann=tf.keras.models.Sequential()
```

```python
ann.add(tf.keras.layers.Dense(units=4,activation='relu'))
```

```python
ann.add(tf.keras.layers.Dense(units=4,activation='relu'))
```

```python
ann.add(tf.keras.layers.Dense(units=1,activation='sigmoid'))
```

##Training the ANN

```python
ann.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
ann.fit(X,y,batch_size=32,epochs=1000)
```

##predict xtest

```python

y_pred=ann.predict(xt)
y_pred=(y_pred>0.5)
```

```python
from sklearn.preprocessing import LabelEncoder
ley=LabelEncoder()
y_pred=ley.fit_transform(y_pred)
pred1 = pd.DataFrame(y_pred, columns=['PassengerId']).to_csv('data11.csv')
```
